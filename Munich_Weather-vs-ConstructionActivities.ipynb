*Background*  
   
Construction in Germany typically is brick and mortar, with considerable portions of most buildings using in-situ concrete parts, even for single family homes and smaller building types. Materials used in this method cannot be safely installed at temperatures below +5°C (41 F) or above +30°C (86 F). So this industry is considered weather dependent and seasonal, with activities limited to interior design during cold/bad weather periods. 
   
We would like to find out whether changing climate and new building technologies and building types, such as prefabricated and modular, have an impact on construction.

The format of CSV files provided on open data portal follows typical column layout:   

- **MONATSZAHL** (Monthly aggregate value type) - String denoting the type of value contained in the corresponding value column
- **AUSPRAEGUNG** (Subtype) - String giving extra details/information on the type of record in row
- **JAHR** (Year) - Is the year associated with the record/row in format "YYYY" as integers
- **MONAT** (Month) - Denotes the month for the aggregated value of the row in format "YYYYMM" as a string value
- **WERT** (Value) - The actual value for the row as a floating number without unit. The unit is derived from the desription of the record in fields "Monatszahl" (Monthly aggregate value type) and "Ausprägung" (Subtype).
- **VORJAHRESWERT** (Value Previous Year) - Floating number of the corresponding value for the monthly aggregate in the previous year
- **VERAEND_VORMONAT_PROZENT** (Change over previous month) - Floating number showing the percentage change of current row value over the previous month
- **VERAEND_VORJAHRESMONAT_PROZENT** (Change over previous year) - Floating number showing the percentage change of current row value over the same month in the previous year
- **ZWOELF_MONATE_MITTELWERT** (Twelve months median) - Floating number giving the average value for the previous 12 months

Let's take a look at the content of column "Monatszahl" to see the potential record types in this file:

```python
list( weather.MONATSZAHL.unique() )
```

The column contains the following record types:  

- **Luftfeuchtigkeit** Relative Humidity
- **Lufttemperature** Air temperature; although not explicitly mentioned in file, values will be in degrees centigrade (°C) as per German standard 
- **Niederschlag** Precipitation; measured in Milimeters (mm)
- **Sonnenschein** Sunshine; measured in hours

We will be using the record type "Lufttemperature" for our analysis only and can thus delete all rows for other record types:

```python
weather = weather[weather['MONATSZAHL'] == 'Lufttemperatur']   
```

We have now reduced the row count from 1.892 to 792:

```python
weather.shape[0]
```

Let's see what subtypes are available for temperature records:

```python
list( weather.AUSPRAEGUNG.unique() )
```

The column contains the following subtypes:

- **Höchste Lufttemperature** Maximum air temperature
- **Mittlere Lufttemperatur** Average air temperature
- **Tiefste Lufttemperatur** Minimum air temperature

Although not explicitly included or mentioned in table, all temperature values are in Degree Centigrade (°C) per European standard.

We can delete a few columns as these will not be needed for analysis or plotting:

```python
weather.drop(['MONATSZAHL', 
                 'VORJAHRESWERT', 
                 'VERAEND_VORMONAT_PROZENT', 
                 'VERAEND_VORJAHRESMONAT_PROZENT',
                 'ZWOELF_MONATE_MITTELWERT'], axis=1, inplace=True)
```

Let's see the content for the year column to get an idea of the period covered in the file:

```python
weather.JAHR.unique()
```

The file contains monthly records for the years 2000 until 2021, a period of 22 years.

We will reduce the filset to a 10 year period for now and discard data outside our range 2011-2020:

```python
for y in [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2021]:
    weather = weather[weather['JAHR'] != y ]     
```

We now have monthly records for the years 2010 until 2020. Let's count records again:

```python
weather.shape[0]
```

```python
weather.head()
```

Let's check if there are some NaN (not a number) values in the table:

```python
weather.shape[0] - weather.dropna().shape[0]
```

No NaN in our table. We're lucky... (If you explore the original raw data set, you will find that there are a few records in the year 2021 that do not contain values.)

Let's convert the shape and layout of the data presentation into a simpler format using separate columns for each data type. (I will overwrite the dataframe for this instead of reformatting it. The new dataframe will use English column names.)

```python
data = [] # temp variable to hold list data

# the method i am using here is not very elegant, but gets the job done...
for y in range(2011, 2021):
    for m in range(1,13):
        d = "{}".format(y) + "{}".format(m).zfill(2)
        
        var_min = weather[(weather['MONAT'] == d) & (weather['AUSPRAEGUNG'] == 'Tiefste Lufttemperatur')]['WERT'].values[0]
        var_avg = weather[(weather['MONAT'] == d) & (weather['AUSPRAEGUNG'] == 'Mittlere Lufttemperatur')]['WERT'].values[0]
        var_max = weather[(weather['MONAT'] == d) & (weather['AUSPRAEGUNG'] == 'Höchste Lufttemperatur')]['WERT'].values[0]
        
        data.append([y, d, var_min, var_avg, var_max ])
        
weather = pd.DataFrame(data, columns=['YEAR', 'MONTH', 'T-MIN', 'T-AVG', 'T-MAX'])
weather.head()
```

Let's do a quick plot of the weather data we have obtained for our observation period:

```python
# sns.reset_defaults()
plt.figure(figsize=(10,4)) 

plt.plot(np.arange(len(weather['T-MAX'])), weather['T-MAX'], '-k', linewidth=0.25, color='#263238', alpha=1, label="High") 
plt.plot(np.arange(len(weather['T-MIN'])), weather['T-MIN'], '-k', linewidth=0.25, color='#263238', alpha=1, label="Low") 
plt.plot(np.arange(len(weather['T-AVG'])), weather['T-AVG'], '-k', linewidth=0.5, color='black', alpha=0.6, label="Avg") 

plt.yticks(color='#777777', alpha=0.8)
xtick = list( range(0, 121, 12))
xlabels = []
for i in range(2010, 2021, 1):
    xlabels.append( str(int(i) + 1) )
plt.xticks(xtick, xlabels, color='#777777', alpha=0.8, rotation=90) 
plt.tick_params(axis='x', color='w')
ax=plt.gca() 

plt.gca().fill_between(np.arange(0,len(weather['T-MAX'])), weather['T-MIN'], weather['T-MAX'], facecolor='#7393B3', alpha=0.5) 

plt.title('Outside Air Temperature Band for Munich, Germany, in years 2001-2020', size=14) 
plt.grid(color='#dddddd', alpha=0.4)
plt.axhline(y=30, color='r', linestyle=':', alpha=0.4)
plt.axhline(y=5, color='b', linestyle=':', alpha=0.4)
plt.text(-5,6,"Band of safe\noperating conditions", color='#777777', va='bottom', fontsize='small', rotation=90, alpha=0.5)

ax=plt.gca() 
ax.set_ylabel('Degree Celsius (°C)', color='#777777', alpha=0.8)
ax2 = ax.twinx()
ax2.set_ylabel('Fahrenheit (F)', color='#777777', alpha=0.8)
y1, y2 = ax.get_ylim()
ax2.set_ylim(((y1 * 1.8) + 32), ((y2 * 1.8) + 32))
ax2.figure.canvas.draw()
plt.yticks(color='#777777', alpha=0.8)

plt.box(False)
plt.show()
```

#### Correlations in Economic Data <a class="anchor" id="economic-correlation">

To validate relationships "seen" in our graphs above so far, we can use Pearson Correlation (standard correlation coefficient) functions.  

`DataFrame.corr(method='pearson')` will return the correlation matrix *p*, where
  

| A *p* value in the range of ....    |  indicates a ... correlation |
| ----------- | ----------- |
| 0.9 - 1      | very strong       |
| 0.7 - 0.9   | strong        |
| 0.5 - 0.7   | moderate        |
| 0.3 - 0.5   | weak        |
| 0 - 0.3   | negligible       |

As expected, Orders Entry does not seem to be a correlated to the other values. The reason is simple: An order is booked in the month when it is received but not assigend to or spread out over the time frame that relates to the project span. Construction projects often are not only multi-months projects, but more often multi-year activities. We will ignore this element for our analysis and concentrate on the other three metrics.

```python
print([ economicdata['ORDERS'].corr(economicdata['SALARIES'], method='pearson'),
       economicdata['ORDERS'].corr(economicdata['HOURS'], method='pearson'),
       economicdata['ORDERS'].corr(economicdata['SALES'], method='pearson')])
```

There is a strong relationship between the other values - we can observe this in the graph and the resulting *p* correlation value:  

- Hours worked and salaries paid do relate by nature, with a lag of a months' interval.   

- Hours worked impact the progress of projects and thus influences invoicing. There is a delay between performance (aka finishing a job) and invoicing to be considered, as the industry typically does not bill on a monthly schedule but according to progress and milestones reached. Thus the relationship is still strong but not as the correlation between hours and salaries.

Scatterplots with regression lines illustrate and support these relations:

```python
fig, axes = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(10,4))
fig.suptitle('Correlation of Economic Data Construction Business in Munich 2011-2020', fontsize=14)
sns.set()

# scatterplots with regression lines
sns.regplot(x='HOURS', y='SALARIES', data=economicdata, ax=axes[0], color='#005b96', scatter_kws={'s':5});
sns.regplot(x='HOURS', y='SALES', data=economicdata, ax=axes[1], color='#005b96', scatter_kws={'s':5});
sns.regplot(x='SALES', y='SALARIES', data=economicdata, ax=axes[2], color='#005b96', scatter_kws={'s':5});
axes[0].set(xticklabels=[], yticklabels=[]);
axes[1].set(xticklabels=[], yticklabels=[]);
axes[2].set(xticklabels=[], yticklabels=[]);
axes[0].tick_params(bottom=False, left=False) 
axes[1].tick_params(bottom=False, left=False) 
axes[2].tick_params(bottom=False, left=False) 
axes[0].xaxis.get_label().set_fontsize(8)
axes[1].xaxis.get_label().set_fontsize(8)
axes[2].xaxis.get_label().set_fontsize(8)
axes[0].yaxis.get_label().set_fontsize(8)
axes[1].yaxis.get_label().set_fontsize(8)
axes[2].yaxis.get_label().set_fontsize(8)

# person correlations
p_val = '$p$ = {0:0.4f} '.format( economicdata['HOURS'].corr(economicdata['SALARIES'], method='pearson') )
axes[0].text(200, economicdata.SALARIES.max(), p_val, horizontalalignment='left', fontsize=10, color='dimgrey');
p_val = '$p$ = {0:0.4f} '.format( economicdata['HOURS'].corr(economicdata['SALES'], method='pearson') )
axes[1].text(200, economicdata.SALES.max(), p_val, horizontalalignment='left', fontsize=10, color='dimgrey');
p_val = '$p$ = {0:0.4f} '.format( economicdata['SALES'].corr(economicdata['SALARIES'], method='pearson') )
axes[2].text(economicdata.SALARIES.min()*2.8, economicdata.SALARIES.max(), p_val, horizontalalignment='left', fontsize=10, color='dimgrey');

fig.show()
sns.reset_orig()
```

#### Seasonality <a class="anchor" id="seasonality">

For a graphical analysis of the seasonal changes in hours worked, we can plot the series of the individual years of our 10 year observation period on top of each other. The steep decline over the turn of the year (a.k.a. in december and january of the following year) is already clearly visible from the raw data itself. Using normalized data (right hand side subplot in below graph) gives a clearer picture:

```python
fig, ax = plt.subplots(1, 2, sharex=False, sharey=False, figsize=(12,4))
fig.suptitle('Hours worked in Months of Year', fontsize=16)

months = list(range(1,13,1))

for y in list(range(2011,2021,1)):
    df = economicdata[economicdata['YEAR']  == y]
    ax[0].plot(months, df['HOURS'], label = y)
    
for y in list(range(2011,2021,1)):
    df = economicdata[economicdata['YEAR']  == y].copy()
    # NORMALIZE USING Z-SCORE
    df['HOURS'] = (df['HOURS'] - df['HOURS'].mean()) / df['HOURS'].std()    
    ax[1].plot(months, df['HOURS'], label = y)

plt.setp(ax, xticks=months, xticklabels=months)   
ax[0].title.set_text('Raw data (Hours in Thousands)')
ax[1].title.set_text('Normalized using Z-Score Method')

mplcursors.cursor(hover=True)
ax[1].text(12, -2.3, 'Hover over line to show value and year', horizontalalignment='right', verticalalignment='center', color='grey', alpha=0.8)

plt.show()
```

Although four out of ten years do show a strong correlation, we see quite a fluctuation overall and cannot identify a clear pattern. Particularly, we can not see the anticipated increase in correlation that could have been expected under the initial premise.

Let us take a closer look at the two years with extremes, i.e. 2017 with a strong correlation of $p$=0.726 and 2014 with a weak correlation of $p$=0.490.

When analysing the amount of hours worked in a given month, we need to consider the impact of reduced working day count in a month due to national holidays and "bridge days" associated. Please note that Germany has rather strict regulations governing working days for most industries including overtime and extra pay requirements. Thus, 24/7 activities on construction sites are expensive and consequently rather rare for the bulk of building projects. The non-working business days are listed in a section next to the year and the number of remaining regular working days in the month is shown in bottom of graph.

```python
def get_df(year):
    df = economicdata[economicdata['YEAR']  == year].copy()
    df.drop(['YEAR', 'ORDERS', 'SALARIES', 'SALES'], axis=1, inplace=True)
    df['HOURS'] = (df['HOURS'] - df['HOURS'].mean()) / df['HOURS'].std()
    w_df = weather[weather['YEAR']  == year].copy()
    w_df['T-AVG'] = (w_df['T-AVG'] - w_df['T-AVG'].mean()) / w_df['T-AVG'].std() 
    df['TEMP'] = w_df['T-AVG']
    df['MONTH'] = df['MONTH'].str[4:]
    df = df.apply(pd.to_numeric, errors='coerce')
    return df

def get_corr(year):
    p = "$p$ = {:5f}".format( economicdata[economicdata['YEAR']  == year]['HOURS'].corr(weather[weather['YEAR']  == year]['T-AVG'], method='pearson') )  
    return p



# main():

# plot layout & init
fig, axes = plt.subplots(2, 1, sharex=False, sharey=False, figsize=(14,8))
fig.suptitle('Hours worked & Average Temperature in Months of Year (Normalized)', fontsize=14)

months = list(range(1,13,1))
xt = list(range(1,13,1))
xtl = [ "JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC" ]


# create plot for 2014
y=2014
df = get_df(y)
line1LHS, line2LHS, = axes[0].plot(df.MONTH, df.HOURS, 'r', df.MONTH, df.TEMP, 'b')
line1LHS.axes.set_xlim(.50, 12.2)
line1LHS.axes.set_ylim(-2.5, 2.5)
line1LHS.axes.set_xticks(xt)
line1LHS.axes.set_xticklabels(xtl)
line1LHS.axes.set_yticks([])
line1LHS.axes.set_yticklabels([])  
tcorrLHS = axes[0].text(0.7, 2.2, get_corr(y), fontsize=10, horizontalalignment='left', verticalalignment='center', color='black', alpha=0.8)
tyearLHS = axes[0].text(12.1, 2.2, y, fontsize=12, fontweight='bold', horizontalalignment='right', verticalalignment='center', color='black', alpha=0.9)
axes[0].grid(False)


# create plot for 2017
y=2017
df = get_df(y)
line1RHS, line2RHS, = axes[1].plot(df.MONTH, df.HOURS,'r', df.MONTH, df.TEMP, 'b')
line1RHS.axes.set_xlim(.50, 12.2)
line1RHS.axes.set_ylim(-2.5, 2.5)
line1RHS.axes.set_xticks(xt)
line1RHS.axes.set_xticklabels(xtl)
line1RHS.axes.set_yticks([])
line1RHS.axes.set_yticklabels([])  
tcorrRHS = axes[1].text(0.7, 2.2, get_corr(y), fontsize=10, horizontalalignment='left', verticalalignment='center', color='black', alpha=0.8)
tyearRHS = axes[1].text(12.1, 2.2, y, fontsize=12, fontweight='bold', horizontalalignment='right', verticalalignment='center', color='black', alpha=0.9)
axes[1].grid(False)


# custom legend
axes[0].text(0.5,3.2, r'$\blacksquare$ Hours', fontsize=10, color='r', rotation=0, ha='left', va='center')
axes[0].text(0.5,2.9, r'$\blacksquare$ Average Temperature', fontsize=10, color='b', rotation=0, ha='left', va='center')
axes[0].text(5.5,2.9, r'$\blacksquare$ (Shaded areas mark periods where trend direction of lines do not correlate)', fontsize=10, color='grey', rotation=0, ha='center', va='center')
axes[0].text(12.15,2.9, r'$\blacksquare$ (Number of Business Days in resp. Month)', fontsize=10, color='green', rotation=0, ha='right', va='center')


# trend diversion indicators
# 2014: APR-JUN, AUG-OCT
rectA1=mpatches.Rectangle((4, -2.5), 2, 5, fill = True, color = "grey", alpha=0.05, linewidth = .1, zorder=-1)
rectA2=mpatches.Rectangle((8, -2.5), 2, 5, fill = True, color = "grey", alpha=0.05, linewidth = .1, zorder=-1)
axes[0].add_patch(rectA1)
axes[0].add_patch(rectA2)
# 2017: MAY-JUN, AUG-NOV
rectB1=mpatches.Rectangle((5, -2.5), 1, 5, fill = True, color = "grey", alpha=0.05, linewidth = .1, zorder=-1)
rectB2=mpatches.Rectangle((8, -2.5), 3, 5, fill = True, color = "grey", alpha=0.05, linewidth = .1, zorder=-1)
axes[1].add_patch(rectB1)
axes[1].add_patch(rectB2)


# number of working days in months for 2014
for i, val in enumerate([21,20,23,18,21,20,21,22,21,20,21,19]):
    axes[0].text(i+1,-2.35, str(val), fontsize=8, color='green', ha='center')
# number of working days in months for 2017
for i, val in enumerate([21,20,21,20,20,19,23,20,22,22,20,21]):
    if i == 0:
        axes[1].text(0.9,-2.35, str(val), fontsize=8, color='green', ha='center') # move to left
    else:
        axes[1].text(i+1,-2.35, str(val), fontsize=8, color='green', ha='center')


# add special event details
header = "National Holidays\non Business Day:" + "\n\n"
notes2014 = "January 1st" + "\n" + \
            "January 6th" + "\n" + \
            "April 18th" + "\n" + \
            "April 21st" + "\n" + \
            "May 1st" + "\n" + \
            "May 29th" + "\n" + \
            "June 9th" + "\n" + \
            "June 19th" + "\n" + \
            "August 15th" + "\n" + \
            "October 3rd" + "\n" + \
            "December 25th" + "\n" + \
            "December 26th"
notes2017 = "January 6th" + "\n" + \
            "April 14th" + "\n" + \
            "April 17th" + "\n" + \
            "May 1st" + "\n" + \
            "May 25th" + "\n" + \
            "June 5th" + "\n" + \
            "June 15th" + "\n" + \
            "August 15th" + "\n" + \
            "October 3rd" + "\n" + \
            "October 31st" + "\n" + \
            "November 1st" + "\n" + \
            "December 25th" + "\n" + \
            "December 26th"
textHeader = axes[0].text(12.3, 3.2, header, fontsize=10, fontstyle='italic', horizontalalignment='left', verticalalignment='top', color='black', alpha=0.7)
text2014 = axes[0].text(12.3, 2.5, notes2014, fontsize=10, horizontalalignment='left', verticalalignment='top', color='black', alpha=0.7)
text2017 = axes[1].text(12.3, 2.5, notes2017, fontsize=10, horizontalalignment='left', verticalalignment='top', color='black', alpha=0.7)



# show the plots
plt.show()
```
